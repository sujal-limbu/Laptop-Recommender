```python
import pandas as pd

df = pd.read_csv("C:\\Users\\limbu\\Laptop_recommendation\\data\\raw\\laptop.csv")
df.head()
```

```python
# Remove ₹, commas, and convert to numeric
df['Price'] = df['Price'].str.replace('₹', '').str.replace(',', '').astype(float)

# Remove 'GB' and convert RAM and SSD to int
df['Ram_GB'] = df['Ram'].str.extract(r'(\d+)').astype(int)
df['SSD_GB'] = df['SSD'].str.extract(r'(\d+)').astype(float).fillna(0)

# Clean display size
df['Display_Inches'] = df['Display'].str.extract(r'(\d+\.?\d*)').astype(float)

# Clean rating (some might be NaN or float strings)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Drop or fix missing values
df.dropna(subset=['Price', 'Ram_GB', 'SSD_GB', 'Display_Inches'], inplace=True)
```

```python
# Convert all to lowercase and strip
for col in ['Generation', 'Core', 'Graphics', 'OS', 'Warranty']:
    df[col] = df[col].astype(str).str.lower().str.strip()
```

```python
# Remove duplicate models (keep the first one)
df.drop_duplicates(subset='Model', inplace=True)

# Optional: drop irrelevant columns (like Unnamed: 0)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
# Reorder and rename for clarity
df = df[['Model', 'Price', 'Rating', 'Generation', 'Core',
         'Ram_GB', 'SSD_GB', 'Display_Inches', 'Graphics', 'OS', 'Warranty']]

# Save cleaned version
df.to_csv("cleaned_laptop_data.csv", index=False)
```
